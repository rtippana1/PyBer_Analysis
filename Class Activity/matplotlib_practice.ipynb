```python
#Import dependencies
import matplotlib.pyplot as plt
import matplotlib.colors
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot using object method.
fig,ax = plt.subplots()
ax.plot(x_axis, y_axis,marker = "D",color = "green",linewidth = 2, label ='Boston')
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add a grid.
ax.grid()
# Add the legend.
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

#Create a scatter plot using plt.scatter method
plt.scatter(y_axis,x_axis,c="red",label ='Chicago')
#Add axes
plt.xlabel("Dates")
plt.ylabel("Fare($)")
#Set limit on X-axis
plt.xlim(0,45)
#Inverse y axis
plt.gca().invert_yaxis()
#Add Title
plt.title("PerBer Fare By Month")
#Add Legend
plt.legend()
```

```python
#Creating a bubble chart
#using list comprehension to increse the size of bubbles which is the third parameter passed to the function
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
#Create the bubble chart using Object Oriented Method'''alpha = 0.20,'''
fig,ax = plt.subplots()
ax.scatter(y_axis,x_axis,c="skyblue",edgecolors="black",alpha = 0.9,s=[i*5 for i in y_axis],label ='Boston',linewidths = 2)
#Add limit to x-axis
ax.set_xlim(0,50)
#inverse the y-axis data
ax.invert_yaxis()
#Add axis title
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
#Add Title
ax.set_title("PerBer Fare By Month")
#Add Legend
ax.legend()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
#Create a PIE chart using OOI
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,explode_values,labels=x_axis)
plt.show()
```
